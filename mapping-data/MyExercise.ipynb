# Load data using Pandas

```python
import pandas as pd

wines_df = pd.read_csv('sample_data/wine-ratings-small.csv', index_col=0)
wines_df.head()
```

# Using iteration to classify wine type

```python
# retrieve all wine variety in the dataset
wines_type = set(wines_df['variety'])
wines_type = list(wines_type)
wines_type
```

```python
#wine_type_dict_name = []
#for i in wine_type:
#    i = str(i)
#    wine_type_dict_name.append(i.replace(" ", ""))


# Create 6 data frame for each wine type
Pink = pd.DataFrame()
WhiteWine = pd.DataFrame()
SparklingChampagne = pd.DataFrame()
RedWine = pd.DataFrame()
Collectible = pd.DataFrame()
nan = pd.DataFrame()


#classify
for i in wines_df.index:
    if wines_df['variety'][i] == 'Pink and Rosé':
        Pink = Pink.append(pd.DataFrame(wines_df.loc[[i]]))
    else:
        if wines_df['variety'][i] == 'Red Wine':
            RedWine = RedWine.append(pd.DataFrame(wines_df.loc[[i]]))
        else:
            if wines_df['variety'][i] == 'White Wine':
                WhiteWine = WhiteWine.append(pd.DataFrame(wines_df.loc[[i]]))
            else:
                if wines_df['variety'][i] == 'Sparkling&Champagne':
                   SparklingChampagne = SparklingChampagne.append(pd.DataFrame(wines_df.loc[[i]]))
                else:
                    if wines_df['variety'][i] == 'Collectible':
                        RCollectiblee = Collectible.append(pd.DataFrame(wines_df.loc[[i]]))
                    else:
                        nan = nan.append(pd.DataFrame(wines_df.loc[[i]]))
        
RedWine
```

### save each wine type to a formatted new JSON file

```python
import json

p = Pink.to_dict()
with open ('JsonFile\Type_Pink.json', "w") as file:
    json.dump(p,file, sort_keys= True, indent=4)

r = RedWine.to_dict()
with open ('JsonFile\Type_Red_Wine.json', "w") as file:
    json.dump(r,file, sort_keys= True, indent=4)

# other types are similar
```
